## Combine datasets

```python
# Create color array for both dataframes using NumPy 
color_red = np.repeat('red', red_df.shape[0])
color_white = np.repeat('white', white_df.shape[0])
```

```python
# Add arrays to the red and white dataframes
red_df['color'] = color_red
white_df['color'] = color_white
```

```python
# Check dataframes
red_df.head()
```

```python
# Check dataframes
white_df.head()
```

```python
# Rename the column name to match white wine columns
red_df = red_df.rename(columns={'total_sulfur-dioxide': 'total_sulfur_dioxide'})
```

```python
# append dataframes
wine_df = red_df.append(white_df, ignore_index=True, sort=False)
# view dataframe to check for success
wine_df.head()
```

## Save Combined Dataset

```python
wine_df.to_csv('Data/winequality_edited.csv', index=False)
```
